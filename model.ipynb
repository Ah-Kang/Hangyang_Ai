```python
import joblib

print("1. 냉동 보관된 AI의 뇌(pkl 파일)를 깨우는 중...")
# 수정 완료: model.load -> joblib.load 로 변경했습니다!
my_ai = joblib.load('LUAD_Prediction_AI_Model.pkl')

print("\n========================================")
print("🧠 [AI 모델 해동 완료 및 정체 확인]")
print("========================================")
# AI가 잘 깨어났는지 스펙을 물어봅시다.
print(f"1) 모델의 종류: {type(my_ai).__name__}")
print(f"2) AI가 배운 유전자(시험과목) 개수: {my_ai.n_features_in_}개")
print(f"3) AI가 내릴 수 있는 진단 종류: {my_ai.classes_}")
print(f"4) AI 안에 있는 결정 나무(의사)의 수: {my_ai.n_estimators}명")
print("========================================")
```
